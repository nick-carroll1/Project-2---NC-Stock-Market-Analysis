```python
# Import the necessary packages
import numpy as np
import random
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Plot the risk and return of the portfolios
def plotPortfolios(portfolioDF):
    portfolioDF.plot.scatter(x = "Expected Risk", y = "Average Return", figsize = (10, 8), xlabel = "Expected Risk", ylabel = "Average Return", title = "Risk and Return of Portfolios")
```

```python
stocks = ['AAPL', 'GOOG', 'MSFT']
```

```python
import marketanalysis as ma
```

```python
# Plot the risk and return of all portfolios
analysis = ma.portfolioAnalysis(stocks)
plotPortfolios(analysis)
```

```python
# Plot the risk and return of the portfolios near the efficient frontier
top15 = ma.efficientFrontier(analysis)
plotPortfolios(top15)
```

```python
import altair as alt
plot = alt.Chart(analysis).mark_point().encode(
    x=alt.X("Expected Risk", scale=alt.Scale(zero=False)), y='Average Return', tooltip="Portfolio Make-up").interactive()
plotTop = alt.Chart(top15).mark_point().encode(x=alt.X(
    "Expected Risk", scale=alt.Scale(zero=False)), y='Average Return', color=alt.value('red'))
plot + plotTop
```
